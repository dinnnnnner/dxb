```python
from __future__ import annotations

import sys
from pathlib import Path

import pandas as pd
from IPython.display import display
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots

if (Path.cwd() / "wheel_cog_outputs").exists():
    sys.path.insert(0, str(Path.cwd() / "wheel_cog_outputs"))

from process_wheel_cog import PROJECT_ROOT

BATCH_OUT_DIR = PROJECT_ROOT / "wheel_cog_outputs" / "fast_alarm_batch_outputs"
DISPLAY_HTML_DIR = BATCH_OUT_DIR / "display_html_algorithm_only"
SUMMARY_CSV = BATCH_OUT_DIR / "fast_alarm_batch_summary.csv"

DISPLAY_HTML_DIR.mkdir(parents=True, exist_ok=True)
pio.renderers.default = "notebook_connected"

summary = pd.read_csv(SUMMARY_CSV)
display_cols = [
    "case_name",
    "frames",
    "alarm_frames",
    "alarm_segments",
    "first_alarm_time_s",
    "last_alarm_time_s",
]
summary[display_cols]
```

```python
def alarm_segments(df: pd.DataFrame) -> list[tuple[float, float]]:
    if "alarm" not in df.columns or df.empty:
        return []
    segments: list[tuple[float, float]] = []
    active = df["alarm"].astype(bool).to_numpy()
    times = df["time_s"].to_numpy()
    start = None
    prev = None
    for is_alarm, t in zip(active, times):
        if is_alarm and start is None:
            start = float(t)
        if not is_alarm and start is not None:
            segments.append((start, float(prev)))
            start = None
        prev = t
    if start is not None and prev is not None:
        segments.append((start, float(prev)))
    return segments


def downsample(df: pd.DataFrame, max_points: int = 6000) -> pd.DataFrame:
    step = max(1, len(df) // max_points)
    return df.iloc[::step].copy()


def plot_case(case_index: int = 0) -> go.Figure:
    row = summary.iloc[case_index]
    alarm_df = pd.read_csv(row["alarm_output_file"], low_memory=False, dtype={"alarm_wheel": "string"})
    plot_df = downsample(alarm_df)
    labels_by_file = {}
    file_name = Path(row["input_file"]).name

    fig = make_subplots(
        rows=3,
        cols=1,
        shared_xaxes=True,
        vertical_spacing=0.04,
        row_heights=[0.52, 0.24, 0.24],
        subplot_titles=("wheel speed", "innovation and thresholds", "evidence"),
    )

    for idx in range(4):
        fig.add_trace(
            go.Scatter(
                x=plot_df["time_s"],
                y=plot_df[f"wheel{idx}_rad_s"],
                mode="lines",
                name=f"wheel{idx}",
                line={"width": 1},
            ),
            row=1,
            col=1,
        )

    fig.add_trace(
        go.Scatter(x=plot_df["time_s"], y=plot_df["innovation"], mode="lines", name="innovation"),
        row=2,
        col=1,
    )
    fig.add_trace(
        go.Scatter(x=plot_df["time_s"], y=plot_df["enter_threshold"], mode="lines", name="enter", line={"dash": "dash"}),
        row=2,
        col=1,
    )
    fig.add_trace(
        go.Scatter(x=plot_df["time_s"], y=-plot_df["enter_threshold"], mode="lines", name="-enter", line={"dash": "dash"}),
        row=2,
        col=1,
    )
    fig.add_trace(
        go.Scatter(x=plot_df["time_s"], y=plot_df["evidence"], mode="lines", name="evidence"),
        row=3,
        col=1,
    )
    fig.add_trace(
        go.Scatter(x=plot_df["time_s"], y=plot_df["on_threshold"], mode="lines", name="evidence_on", line={"dash": "dash"}),
        row=3,
        col=1,
    )

    for start_s, end_s in alarm_segments(alarm_df):
        fig.add_vrect(
            x0=start_s,
            x1=end_s,
            fillcolor="#d62728",
            opacity=0.24,
            line_width=1,
            line_color="#d62728",
            annotation_text="algorithm alarm",
            annotation_position="top right",
        )



    fig.update_layout(
        title=f"{case_index}: {row['case_name']} - algorithm alarm only",
        height=850,
        hovermode="x unified",
        dragmode="pan",
        legend={"orientation": "h", "yanchor": "bottom", "y": 1.02, "xanchor": "left", "x": 0},
    )
    fig.update_xaxes(title_text="time_s", row=3, col=1)
    fig.update_xaxes(rangeslider={"visible": True}, row=3, col=1)
    fig.update_yaxes(title_text="rad/s", row=1, col=1)
    return fig


summary[["case_name", "input_file"]]
```

```python
PLOT_CONFIG = {
    "scrollZoom": True,
    "displaylogo": False,
    "responsive": True,
    "modeBarButtonsToAdd": ["drawline", "eraseshape"],
}

figures = []
html_files = []
# for CASE_INDEX in range(len(summary)):
#     fig = plot_case(CASE_INDEX)
#     html_path = DISPLAY_HTML_DIR / f"{CASE_INDEX:02d}_{summary.iloc[CASE_INDEX]['case_name']}.html"
#     fig.write_html(html_path, include_plotlyjs="cdn", config=PLOT_CONFIG)
#     html_files.append(str(html_path))
#     fig.show(config=PLOT_CONFIG)
#     figures.append(fig)
CASE_INDEX =8
fig = plot_case(CASE_INDEX)
html_path = DISPLAY_HTML_DIR / f"{CASE_INDEX:02d}_{summary.iloc[CASE_INDEX]['case_name']}.html"
fig.write_html(html_path, include_plotlyjs="cdn", config=PLOT_CONFIG)
html_files.append(str(html_path))
fig.show(config=PLOT_CONFIG)
figures.append(fig)


pd.DataFrame({"html_file": html_files})
```
