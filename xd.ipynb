```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# 1. 定义共形变换函数
# =========================
def f(z):
    # 示例1：平方变换
    # return z**2

    # 示例2：反演变换
    # return 1 / z

    # 示例3：Mobius变换
    # return (z - 1) / (z + 1)

    # 示例4：指数变换
    return np.exp(z)


# =========================
# 2. 生成原始复平面网格
# =========================
x = np.linspace(-3, 3, 25)
y = np.linspace(-3, 3, 25)

t = np.linspace(-3, 3, 500)

# =========================
# 3. 绘图
# =========================
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

ax1, ax2 = axes

# 原始平面
for xi in x:
    z = xi + 1j * t
    ax1.plot(z.real, z.imag, 'b', linewidth=0.6)

for yi in y:
    z = t + 1j * yi
    ax1.plot(z.real, z.imag, 'r', linewidth=0.6)

ax1.set_title("Original z-plane")
ax1.set_xlabel("Re(z)")
ax1.set_ylabel("Im(z)")
ax1.axis("equal")
ax1.grid(True)


# 变换后平面
for xi in x:
    z = xi + 1j * t

    # 避免 Mobius 变换中 z = -1 的奇点
    z = z[np.abs(z + 1) > 1e-3]

    w = f(z)
    ax2.plot(w.real, w.imag, 'b', linewidth=0.6)

for yi in y:
    z = t + 1j * yi
    z = z[np.abs(z + 1) > 1e-3]

    w = f(z)
    ax2.plot(w.real, w.imag, 'r', linewidth=0.6)

ax2.set_title("Mapped w-plane")
ax2.set_xlabel("Re(w)")
ax2.set_ylabel("Im(w)")
ax2.axis("equal")
ax2.grid(True)

plt.tight_layout()
plt.show()
```
